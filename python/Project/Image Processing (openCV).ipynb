```python
import cv2
i1=cv2.imread("C:\\Users\\arnab\\Downloads\\thor.jpg",1)
i1=cv2.resize(i1,(1080,700))
i1=cv2.flip(i1,1)
cv2.imshow("thor",i1)
p=cv2.waitKey(0)
if(p==ord('q')):
    cv2.imwrite("C:\\Users\\arnab\\Downloads\\thor2.png",i1)
    cv2.destroyAllWindows()
else:
    cv2.destroyAllWindows()
```
