```python
import cv2
vid_cap=cv2.VideoCapture("C:\\Users\\arnab\\Downloads\\Sharp & Crisp PORSCHE.mp4")
print (vid_cap)           
while True:
    r,frame=vid_cap.read()                           #vid_cap.read returns r=boolean value(true,false) & frame=image
    frame=cv2.resize(frame,(480,720))
    cv2.imshow("video",frame)
    k=cv2.waitKey(25)
    if k==ord("q"):
        break
vid_cap.release()
cv2.destroyAllWindows() 
```

```python
#to convert a video in gray 
import cv2
vid_cap=cv2.VideoCapture("C:\\Users\\arnab\\Downloads\\Sharp & Crisp PORSCHE.mp4")
print (vid_cap)           
while True:
    r,frame=vid_cap.read()                           #vid_cap.read returns r=boolean value(true,false) & frame=image
    frame=cv2.resize(frame,(480,720))
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow("Gray Video",gray)
    cv2.imshow("video",frame)
    k=cv2.waitKey(20)
    if k==ord("q"):
        break
vid_cap.release()
cv2.destroyAllWindows() 
```

```python
#accessing video through webcam
import cv2
cap=cv2.VideoCapture(0)                                           #0=for internal camera, 1= extrenal camera
#cap=cv2.VideoCapture(0,cv2._CAP_DSHOW)                           # (i) this is used if any warning or error occurs
print(cap)
while cap.isOpened():
    r,frame=cap.read()          
    if r==True:                                                   #untill the webcam is opened, it returns true value
        frame=cv2.resize(frame,(700,500))
        gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        cv2.imshow("Gray Video",gray)
        cv2.imshow("video",frame)
        k=cv2.waitKey(20)
        if k==ord("q"):                                           # (ii) if error occurs then use:  if k==ord("q")& 0XFF:
            break
cap.release()
cv2.destroyAllWindows() 
```

```python
#accessing flipped video through webcam
import cv2
cap=cv2.VideoCapture(0)                                                                    
print(cap)
while cap.isOpened():
    r,frame=cap.read()          
    if r==True:                                                   
        frame=cv2.resize(frame,(700,500))
        frame=cv2.flip(frame,0)                                   #for flipping
        gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        cv2.imshow("Gray Video",gray)
        cv2.imshow("video",frame)
        k=cv2.waitKey(20)
        if k==ord("q"):                                           
            break
cap.release()
cv2.destroyAllWindows() 
```

```python
#accessing & saving the video from webcam to desired location
import cv2
cap=cv2.VideoCapture(0)    
f=cv2.VideoWriter_fourcc(*'XVID')   #format for accesssing video through webcam- XVID,WMV2,MJPG,WMV1,X264     
output=cv2.VideoWriter("C:\\Users\\arnab\\Downloads\\video_from_python.mp4",f,20.0,(720,480))      #"location",access_name,fps,resolution
print(cap)
while cap.isOpened():
    r,frame=cap.read()          
    if r==True:                                                   
        frame=cv2.resize(frame,(720,480))
        gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        cv2.imshow("Gray Video",gray)
        cv2.imshow("video",frame)
        output.write(frame)
        k=cv2.waitKey(20)
        if k==ord("q"):                                           
            break
cap.release()                 #after the task executed, releases everything to avoid errors
output.release()
cv2.destroyAllWindows() 
```
