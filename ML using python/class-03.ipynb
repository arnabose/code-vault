```python
a=np.random.randint(18,45,(100))        #it will print 100 random numbers between 18 to 45 (int is used to take only integers)
```

```python
plt.hist(a)
plt.show()
```

```python
plt.hist (a)
plt.title("plot histogram")
plt.xlabel("value of a")
plt.ylabel("no. of occurance")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(a,bins)
plt.title("plot histogram")
plt.xlabel("value of a")
plt.ylabel("no. of occurance")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(a,bins,width=0.8)
plt.title("plot histogram")
plt.xlabel("value of a")
plt.ylabel("no. of occurance")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(a,bins, histtype="step")
plt.title("plot histogram")
plt.xlabel("value of a")
plt.ylabel("no. of occurance")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(a,bins,rwidth=0.8, histtype="bar", orientation="horizontal")
plt.title("plot histogram")
plt.xlabel("value of a")
plt.ylabel("no. of occurance")
plt.show()
```

```python
from matplotlib import style
#style.use("ggplot")
plt.figure(figsize=(16,9))       #changes the size(ratio,ratio)
plt.hist([a,b],bins,rwidth=0.8,histtype="bar",color=["m","y"],label=["a","b"])
plt.title("histogram")
plt.xlabel("value of a and b")
plt.ylabel("no. of occurance")
plt.legend()
plt.show()
```
