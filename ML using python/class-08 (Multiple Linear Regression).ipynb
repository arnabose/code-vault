```python
#feature scaling
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
sc.fit(x_train)
```

```python
x_train=sc.transform(x_train)    #using standard scaler algo, 
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train) 
```

```python
lr.coef_                 #used to show the value of m (slope)
```

```python
lr.intercept_              #used to find the value of constatnt term C
```

```python
x_test[0,:]              #[row,column]     here 0 represents thr index of row
```

```python
lr.predict([x_test[0,:]])
```

```python
lr.predict(x_test)
```

```python
y_test
```

```python
lr.score(x_test,y_test)
```

# Lasso and Ridge algorithm

```python
from sklearn.linear_model import Ridge,Lasso
```

```python
rd=Ridge()
```

```python
rd.fit(x_train,y_train)
```

```python
rd.score(x_test,y_test)
```

```python
ld=Lasso()
```

```python
ld.fit(x_train,y_train)
```

```python
ld.score(x_test,y_test)
```

```python
rd1=Ridge(alpha=2)
rd1.fit(x_train,y_train)
rd1.score(x_test,y_test)
```

```python
ld1=Lasso(alpha=2)
ld1.fit(x_train,y_train)
ld1.score(x_test,y_test)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error   #used to find the error of actual value and predicted value
```

```python
mse=mean_squared_error(y_test,y_pred)    #mean square error
```
