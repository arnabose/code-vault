```python
a=np.arange(1,13)           #used to create array based on numerical values (from, to-1, interval)
a
```

```python
a1=np.arange(1,20,3)
a1
```

```python
a2=np.linspace(1,10,4)                #equally distributes the difference
a2
```

```python
a3=a.reshape(2,6)              #it is used to change the shape of an array without changing its data. (no. of array,row,column)
a3
```

```python
a4=a.reshape(2,3,2)
a4
```

```python
a4.ravel()             #used to change a multi-dimensional array into a contiguous flattened array
```
