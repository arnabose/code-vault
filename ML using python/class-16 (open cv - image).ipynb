```python
img1=cv2.imread("C:\\Users\\arnab\\Pictures\\Camera Roll\\arnab.jpeg",1)    # 1= default
img2=cv2.imread("C:\\Users\\arnab\\Pictures\\Camera Roll\\arnab.jpeg",-1)   # -1= increases saturation
img1=cv2.resize(img1,(520,700))
img2=cv2.resize(img2,(520,700))
cv2.imshow("arnab-original image",img1)
cv2.imshow("arnab-saturated",img2)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#image flip
img1=cv2.imread("C:\\Users\\arnab\\Pictures\\Camera Roll\\arnab.jpeg")
img1=cv2.resize(img1,(400,700))
img_f=cv2.flip(img1,-1)                       #1=mirror image,  0=upside down, -1= upside down mirror image
cv2.imshow("arnab",img_f)
cv2.waitKey(3000)
cv2.destroyAllWindows()
```

```python
#user will provide a path and after entering "q", flipped gray image will be saved in download location
path=input("Enter path: ")
print(path)
img=cv2.imread(path,0)
img=cv2.resize(img,(480,720))
img_f=cv2.flip(img,1)
cv2.imshow("original image",img)
cv2.imshow("flipped image",img_f)
k=cv2.waitKey(0)
if k==ord("q"):                                               #ord converts alfabet to ascii value
    cv2.imwrite("C:\\Users\\arnab\\Downloads\\new_arnab_img.png",img_f)
else:
    cv2.destroyAllWindows()
```
