```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer()
```

```python
data.data
```

```python
data.feature_names              #to show the features present in data
```

```python
data.target                       #to show the targeted values (malignant=0, benign=1)
```

```python
 data.target_names                       #to show the target names
```

```python
df=pd.DataFrame(np.c_[data.data,data.target],columns=[list(data.feature_names)+['target']])    
```

```python
df
```

```python
df.shape
```

```python
x=df.iloc[:,0:-1]
```

```python
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=2020)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)
```

Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mnb=MultinomialNB()
```

```python
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test)
```

Bernoulli Naive Bayes

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
bnb=BernoulliNB()
```

```python
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test)
```
