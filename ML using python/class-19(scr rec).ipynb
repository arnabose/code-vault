```python
#screen recording
import cv2 as c
import pyautogui as p
import numpy as np
rs=p.size()                                      #creates resolution
fn=input("Enter path: ")
f1=c.VideoWriter_fourcc(*'XVID')     
output=c.VideoWriter(fn,f1,20.0,rs) 
#creating recording module
c.namedWindow("Arnab live",c.WINDOW_NORMAL)
c.resizeWindow("Arnab live",(720,480))           #("name",resolution)
while True:
    img=p.screenshot()
    f=np.array(img)
    f=c.cvtColor(f,c.COLOR_BGR2RGB)
    output.write(f)
    c.imshow("Arnab Live",f)
    if c.waitKey(1)==ord("q"):
        break
output.release()
c.destroyAllWindows()
```

```python
#Breaking video into multiple images
import cv2
cap=cv2.VideoCapture("C:\\Users\\arnab\\Downloads\\Sharp & Crisp PORSCHE.mp4")
r,frame=cap.read()
count=0
while True:
    if r==True:
        cv2.imwrite("C:\\Users\\arnab\\Downloads\\images\\img%d.jpg"%count,frame)       #images from video saved here
        cap.set(cv2.CAP_PROP_POS_MSEC,(count**100))                                     #predefined package used to capture images to count^1000 ms
        r,frame=cap.read()
        count=count+1
        cv2.imshow("Sharp & Crisp PORSCHE",frame)
        #print(count)
        if cv2.waitKey(1)==ord("q"):
            break
cap.release()
cv2.destroyAllWindows()
```
